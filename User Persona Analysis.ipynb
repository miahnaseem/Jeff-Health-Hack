## Survey Data Insights

```python
values, counts = np.unique(survey_df['Would you license or sell your health data in order to lower your healthcare costs and/or for profit?'], return_counts=True)
```

```python
normalized_counts = counts/counts.sum()
```

```python
plt.figure(figsize=(8,6))
plt.bar(values, normalized_counts * 100)
plt.xlabel('Sell Data')
plt.ylabel('Count')
plt.xticks(rotation = -60)
sns.despine()
plt.title("Sell Data Distribution")
plt.show()
```

```python
corr_mat = data.corr()
```

```python
plt.figure(figsize=(10,8))
plt.yticks(rotation=45)

mask = np.triu(np.ones_like(corr_mat, dtype=bool))

g = sns.heatmap(corr_mat, mask=mask, cmap='coolwarm', center=0, vmin=-1, vmax=1, annot=True)

g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.savefig('heatmap.png')
plt.show()
```

## Data Transformation for Odds Ratio

```python
table = data[['Socio-Economic Upper', 'Would license and sell data', 'Would not license and sell data']].groupby('Socio-Economic Upper').sum().values
print(table)
```

```python
oddsratio, pvalue = stats.fisher_exact(table)
print("OddsR: ", oddsratio, "p-Value:", pvalue)
```

```python
table = data[['Socio-Economic Middle', 'Would not license and sell data', 'Would license and sell data']].groupby('Socio-Economic Middle').sum().values
print(table)
```

```python
oddsratio, pvalue = stats.fisher_exact(table)
print("OddsR: ", oddsratio, "p-Value:", pvalue)
```

```python
table = data[['Socio-Economic Lower', 'Would license and sell data', 'Would not license and sell data']].groupby('Socio-Economic Lower').sum().values
print(table)
```

```python
oddsratio, pvalue = stats.fisher_exact(table)
print("OddsR: ", oddsratio, "p-Value:", pvalue)
```

```python
table = data[['Would not get DNA sequenced - privacy concerns', 'Would license and sell data', 'Would not license and sell data']].groupby('Would not get DNA sequenced - privacy concerns').sum().values
print(table)
```

```python
oddsratio, pvalue = stats.fisher_exact(table)
print("OddsR: ", oddsratio, "p-Value:", pvalue)
```
